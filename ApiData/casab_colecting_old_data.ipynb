```python
import requests
import pandas as pd
import pyodbc as pyo
from datetime import date
```

```python
# creating a df
df = pd.DataFrame(columns=["id",
                           "created_at",
                           "reference",
                           "pac_values_1",
                           "pac_values_2",
                           "pac_values_3",
                           "pac_values_4",
                           "total_energy_values_1",
                           "total_energy_values_2",
                           "total_energy_values_3",
                           "total_energy_values_4",
                           "year_energy_values_1",
                           "year_energy_values_2",
                           "year_energy_values_3",
                           "year_energy_values_4",
                           "year_energy_unit",
                           ])
```

```python
date = "2022011"
url = 'https://data.exia.com.co/demo/inverter/fronius?date=' + date
response = requests.get(url).json()

for count in response:
    id = count['id']
    created_at = count['created_at']
    #date = str(count['created_at']).split(' ')
    #energyreal_wac_sum_consumed = count['energyreal_wac_sum_consumed']
    #energyreal_wac_sum_produced = count['energyreal_wac_sum_produced']
    #current_ac_sum = count['current_ac_sum']
    #powerreal_p_sum = count['powerreal_p_sum']
    reference = count['reference']
    #voltage_ac_phase_average = count['voltage_ac_phase_average']
    pac_values_1 = count['pac_values_1']
    pac_values_2 = count['pac_values_2']
    pac_values_3 = count['pac_values_3']
    pac_values_4 = count['pac_values_4']
    total_energy_values_1 = count['total_energy_values_1']
    total_energy_values_2 = count['total_energy_values_2']
    total_energy_values_3 = count['total_energy_values_3']
    total_energy_values_4 = count['total_energy_values_4']
    year_energy_values_1 = count['year_energy_values_1']
    year_energy_values_2 = count['year_energy_values_2']
    year_energy_values_3 = count['year_energy_values_3']
    year_energy_values_4 = count['year_energy_values_4']
    year_energy_unit = count['year_energy_unit']
    
    

    #save data into df
    df = df.append({"id": id,
              "created_at": created_at,
              "reference": reference,
              "pac_values_1":pac_values_1,
              "pac_values_2": pac_values_2,
              "pac_values_3": pac_values_3,
              "pac_values_4": pac_values_4,
              "total_energy_values_1":total_energy_values_1,
              "total_energy_values_2": total_energy_values_2,
              "total_energy_values_3": total_energy_values_3,
              "total_energy_values_4": total_energy_values_4,
              "year_energy_values_1": year_energy_values_1,
              "year_energy_values_2": year_energy_values_2,
              "year_energy_values_3": year_energy_values_3,
              "year_energy_values_4": year_energy_values_4,
              "year_energy_unit": year_energy_unit,
                    
              }, ignore_index=True)
```
